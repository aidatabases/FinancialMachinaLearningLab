```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train = pd.read_csv('../input/home-credit-default-risk/application_train.csv')
test = pd.read_csv('../input/home-credit-default-risk/application_test.csv')
```

```python
train.head()
```

```python
feat_names=list(train.columns)
print(feat_names)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['NAME_CONTRACT_TYPE'] = le.fit_transform(train['NAME_CONTRACT_TYPE'])
train['FLAG_OWN_CAR'] = le.fit_transform(train['FLAG_OWN_CAR'])
train['FLAG_OWN_REALTY'] = le.fit_transform(train['FLAG_OWN_REALTY'])

test['NAME_CONTRACT_TYPE'] = le.fit_transform(test['NAME_CONTRACT_TYPE'])
test['CODE_GENDER'] = le.fit_transform(test['CODE_GENDER'])
test['FLAG_OWN_CAR'] = le.fit_transform(test['FLAG_OWN_CAR'])
test['FLAG_OWN_REALTY'] = le.fit_transform(test['FLAG_OWN_REALTY'])
```

```python
train.shape
```

```python
test.shape
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
train.shape
```

```python
test.shape
```

```python
train.columns
```

```python
train_col=set(train.columns)
test_col=set(test.columns)
print(len(train_col))
print(len(test_col))
exc_col = train_col-test_col
print(exc_col)
```

```python
train_target_labels = train['TARGET']
train_align, test_align = train.align(test,join='inner',axis=1)
train_align['TARGET'] = train_target_labels
print('train shape:',train_align.shape)
print('test shape:',test_align.shape)
```

```python
train=train_align
test=test_align
train_label=train.pop('TARGET')
```

```python
print(train_label)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
si=SimpleImputer(missing_values=np.nan,strategy='median')
scaler=MinMaxScaler(feature_range=(0,1))

```

```python
train.shape
```

```python

train=si.fit_transform(train)
test=si.fit_transform(test)
scaler.fit(train)
train = scaler.fit_transform(train)
test = scaler.fit_transform(test)
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_auc_score

x_train, x_test, y_train, y_test = train_test_split(train, train_label, test_size=0.3, random_state=0)

dtree = tree.DecisionTreeClassifier(random_state=50,max_depth=5)
dtree = dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)
#print(len(y_pred))
```

```python
y_pred.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
print('Accuracy Score: {}'.format(accuracy_score(y_test,y_pred)))
print('ROC AUC Score: {}'.format(roc_auc_score(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50,random_state=10, verbose = 1, n_jobs = -1)
rfc.fit(x_train, y_train)

rfc_pred = rfc.predict(x_test)
print(confusion_matrix(y_test,rfc_pred))
print(classification_report(y_test,rfc_pred))
```

```python
print('Accuracy Score: {}'.format(accuracy_score(y_test,rfc_pred)))
print('ROC AUC Score: {}'.format(roc_auc_score(y_test,rfc_pred)))
```
